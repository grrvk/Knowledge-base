```python
from owlready2 import *

lab2 = get_ontology("/content/lab2.rdf").load()

object_properties = lab2.object_properties()
```

```python
#@title Functions

def calculate_total_relations():
    for_indiv = 0
    for prop in list(lab2.properties()):
      for_indiv += len(list(prop.get_relations()))

    total = 0
    for clas in list(lab2.classes()):
      for precise_class in clas.ancestors():
        for prop in precise_class.get_class_properties():
            for value in prop[precise_class]:
              total += 1
    total += for_indiv
    return total

def get_overall_info():
  print(f"Number of classes: {len(list(lab2.classes()))}")
  print(f"Number of individuals: {len(list(lab2.individuals()))}")
  print(f"Number of object properties: {len(list(lab2.object_properties()))}")
  print(f"Number of data properties: {len(list(lab2.data_properties()))}")
  print(f"Number of relations (total): {calculate_total_relations()}")

def get_precise_info():
  for clas in list(lab2.classes()):
    print(f"Class name: {clas.name}")

    print(f"Class ancestors: {clas.ancestors()}")

    if list(clas.subclasses()):
      print(f"Class subclasses: {list(clas.subclasses())}")
    else :
      print(f"Class subclasses: None")

    if len(clas.get_class_properties()) != 0:
      print("Properties:")
      for prop in clas.get_class_properties():
        for value in prop[clas]:
          print(f"{clas.name} {prop.name} {value.name}")
    else: print("Properties: None")

    if len(clas.instances()) != 0:
      print("Individuals:")
      for inst in clas.instances():
        print(f"Individual: {inst.name}")
        if len(list(inst.get_properties())) != 0:
          for prop in list(inst.get_properties()):
            "Properties:"
            for value in prop[inst]:
              print(f"{inst.name} {prop.name} {value}")
        else: print("Properties: none")
    else: print("Individuals: None")

    print('\n')
```
